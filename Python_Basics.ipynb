```python
#6,7.  Write a function to demonstrate the behavior of mutable and immutable arguments
def demonstrate_mutable_immutable(num, lst):
    num += 1
    print("Inside function (immutable):", num)

    lst.append(100)
    print("Inside function (mutable):", lst)

number = 5
my_list = [1, 2, 3]

demonstrate_mutable_immutable(number, my_list)

print("Outside function (immutable):", number)
print("Outside function (mutable):", my_list)
# In this program the number that is an integer will not change out the function rather it will make a copy of it while in function but the list will reflect changes
```

```python
#14. Write a program to reverse a string using a while loop
st = input("Ente a String: ")
rev_st = ""
i = len(st)-1
while i >= 0:
    rev_st = rev_st + st[i]
    i = i-1
print(rev_st)
```

```python
# 15. 0 Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
i = 1
fact = 1
while i <= num:
    fact *= i
    i += 1
print(f"The factorial of the number is: {fact}")
```
